```python
import pandas as pd
data = pd.read_csv("StudentsPerformance.csv")
data.head(10)
```

```python
# Dataset information
data.info()

# Quick numeric summary
data.describe()
```

The dataset contains 1000 rows and 8 columns.
No column shows any missing values.

```python
# Check missing values in each column
data.isnull().sum()
```

"There was no missing value in the dataset"

```python
# Check duplicate rows
print("Duplicate rows before:", data.duplicated().sum())

# Remove duplicates
data.drop_duplicates(inplace=True)

print("Duplicate rows after:", data.duplicated().sum())
```

The dataset contained no duplicate rows.
Therefore, no records were removed during this step.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
sns.boxplot(data=data[['math score','reading score','writing score']])
plt.title("Outlier Check in Student Scores")
plt.show()
```

"A few small outliers were visible in the math score and writing score columns.
However, these values represent real student performance, so no major outliers were removed."

```python
# Save cleaned data
data.to_csv("Cleaned_StudentsPerformance.csv", index=False)
```
