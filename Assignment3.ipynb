```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cleaned dataset load karo
data = pd.read_csv("Cleaned_StudentsPerformance.csv")

data.head()
```

```python
data.groupby('gender')[['math score','reading score','writing score']].mean().plot(kind='bar')
plt.title("Average Scores by Gender")
plt.ylabel("Average Score")
plt.show()
```

Females usually score higher in reading and writing, while males have slightly better math scores.

```python
sns.boxplot(x='lunch', y='math score', data=data)
plt.title("Math Scores Based on Lunch Type")
plt.show()
```

Students with standard lunch score higher than those with free/reduced lunch, which may indicate financial background impact.

```python
sns.boxplot(x='test preparation course', y='writing score', data=data)
plt.title("Effect of Test Preparation on Writing Score")
plt.show()
```

Test preparation course helps improve scores. Students who completed the course perform better.

```python
sns.histplot(data['math score'], kde=True)
plt.title("Math Score Distribution")
plt.show()
```

Most students score around the middle range, showing a normal distribution of performance.

```python
# Sirf numeric columns select karo
numeric_data = data.select_dtypes(include=['int64', 'float64'])

# Ab heatmap banao
sns.heatmap(numeric_data.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Numeric Features Only)")
plt.show()
```

The heatmap shows that reading score and writing score are highly positively correlated, meaning students who read well also tend to write well. Math score has moderate correlation with reading and writing, indicating overall academic consistency.
