```python
import pandas as pd

data = pd.read_csv("Cleaned_StudentsPerformance.csv")   # Agar cleaned file use kar rahi ho
# ya agar original data hai to:
# data = pd.read_csv("StudentsPerformance.csv")

data.head()
```

```python
from statistics import mode

print("Mean Scores:")
print(data[['math score','reading score','writing score']].mean())

print("\nMedian Scores:")
print(data[['math score','reading score','writing score']].median())

print("\nMode Scores:")
print(data[['math score','reading score','writing score']].mode().iloc[0])

print("\nVariance:")
print(data[['math score','reading score','writing score']].var())
```

Mean shows average performance

Median shows middle performance

Mode shows most common score

Variance batata hai kitna variation hai scores me

```python
data['average_score'] = (data['math score'] + data['reading score'] + data['writing score']) / 3
data.head()
```

```python
numeric_data = data[['math score','reading score','writing score','average_score']]
corr = numeric_data.corr()
corr
```

```python
corr_target = corr['average_score'].sort_values(ascending=False)
corr_target
```

The top 3 features most related to student performance are:

Writing Score (Correlation ≈ 0.96)

Reading Score (Correlation ≈ 0.97)

Math Score (Correlation ≈ 0.91)

This means that students who read well also write well and generally perform better academically. Reading and writing abilities are strongly linked to overall performance, while math also contributes but slightly less comparatively.
